```python
import pandas as pd 
import numpy as np 

df = pd.read_csv('../data/netflix_titles.csv')
df.head()
```

```python
# convert the 'date_added' column to datetime format and extract year and month features
df["date_added"] = pd.to_datetime(df["date_added"], errors='coerce')

# Handle missing date_added (fill with median date)
df['date_added'] = df['date_added'].fillna(df['date_added'].median())

df["year_added"] = df["date_added"].dt.year.astype('Int64')
df["month_added"] = df["date_added"].dt.month.astype('Int64') 
```

```python
# content age feature
current_year = pd.Timestamp.now().year
df["content_age"] = current_year - df["release_year"]
```

```python
# primary genre feature
df['primary_genre'] = df['listed_in'].str.split(',').str[0]
```

```python
df['duration_minutes'] = df['duration'].str.extract('(\d+)')
df['duration_minutes'] = pd.to_numeric(df['duration_minutes'], errors='coerce').astype('Int64')
```

```python
# region feature
df['primary_country'] = df['country'].str.split(',').str[0]
```

```python
# binary indicator for movie
df["is_movie"] = np.where(df["type"] == "Movie", 1, 0)
```

```python
df.to_csv("../data/netflix_featured.csv", index=False)
df.head()
```
